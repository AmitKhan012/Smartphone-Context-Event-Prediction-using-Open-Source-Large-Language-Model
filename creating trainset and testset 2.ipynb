```python
df.reset_index(inplace=True)
df = df.drop(columns=['index'], axis=1)
df
```

```python
df = df[df['timestamp'] != '0000-00-00 00:00:00'].copy()
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
dev_id = df['device_id'].unique()
len(dev_id)
```

```python
df2 = df
```

```python
# get device ids where user had device in phone atleast continous 30 days - quinn

# Convert the 'timestamp' column to datetime objects
df2['timestamp'] = pd.to_datetime(df2['timestamp'])

# Sort the DataFrame by 'device_id' and 'timestamp'
df2 = df.sort_values(by=['device_id', 'timestamp'])

# Calculate the difference between the maximum and minimum timestamp for each device
df2['time_range'] = df2.groupby('device_id')['timestamp'].transform('max') - df2.groupby('device_id')['timestamp'].transform('min')

# Define the time difference threshold for continuity (30 days)
continuity_threshold = pd.Timedelta(days=30)

# Filter the DataFrame to get devices where this difference is at least 30 days
at_least_30_days_data = df2[df2['time_range'] >= continuity_threshold]
```

```python
dffinal = at_least_30_days_data.drop(columns=['time_range'], axis=1)
```

```python
dffinal.head()
```

```python
# converting time to seconds
dffinal['timestamp'] = pd.to_datetime(dffinal['timestamp'])
dffinal['timestamp'] = (dffinal['timestamp'] - dt.datetime(1970,1,1)).dt.total_seconds()
```

```python
dffinal.head()
```

```python
actions = dffinal['action'].unique()
len(actions)
```

```python
dffinal = dffinal.drop_duplicates(subset = ['action', 'timestamp'], keep = 'first')
```

```python
dffinal = dffinal.sort_values(by=['device_id', 'timestamp'], ascending=[True, True])
```

```python
dffinal.reset_index(inplace=True)
dffinal = dffinal.drop(columns=['index'], axis=1)
```

```python
dffinal = dffinal.groupby(['device_id', 'timestamp'])['action'].apply(list).reset_index(name = 'action')
```

```python
dffinal
```

```python
minL = len(dffinal['action'].iloc[0])
maxL = len(dffinal['action'].iloc[0])
for i in range(1598949):
    curL = len(dffinal['action'].iloc[i])
    minL = min(minL, curL)
    maxL = max(maxL, curL)
print(minL)
print(maxL)
```

```python
# making dataset
dataset = []

for i in range(1598949):
    if len(dffinal['action'].iloc[i]) > 1:
        dataset.append(dffinal['action'].iloc[i])
```

```python
l = len(dataset)
trainDataset = dataset[:int(0.8*l)]
testDataset = dataset[int(0.8*l):]
```
